```python
# TODO define model
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Convolutional layer with 32 filters, kernel size of 5*5 and stride of 1
        self.conv = nn.Conv2d(3, 32, kernel_size=5, stride=1)
        # Max Pooling layer with kernel size of 2*2 and default stride (kernel_size)
        self.pool = nn.MaxPool2d(kernel_size=2)
        # ReLU Activation Layers
        self.relu = nn.ReLU()
        # Linear layer with output of 512
        self.fc1 = nn.Linear(32*14*14, 512)
        # A linear layer with output of 10.
        self.fc2 = nn.Linear(512, 10)
    
    # format: N x C x W x H
    # input: 256 x 3 x 32 x 32
    # after conv: 256 x 32 x 28 x 28
    # after pool: 256 x 32 x 14 x 14
    # after relu: 256 x 32 x 14 x 14
    # after view: 256 x 6272

    def forward(self, x): 
        x = self.relu(self.pool(self.conv(x)))
        x = x.view(-1, 32*14*14)  # flattens the image
        x = self.relu(self.fc1(x))
        out = self.fc2(x)
        # softmax?
        return out
```

#### **Questions**

- Padding is used in the convolutional layers. What does that mean, which variants exist and how does it influence the dimensions of the feature maps?
- What does padding=1 mean in the case of the network displayed above?

**Your answers here**

*   Padding add a outer layer to output matrix with zero values. Same => gives the matrix with same size of input. Valid => It is other term of no padding. Adding padding gives the pixels on the border of the image more of an opportunity for interacting with the filter
*   Padding = 1 add just 1 zero output layer to matrix.
*   Output = 32 * 32 * 32
*   Output = 32 * 32 * 64
*   16 * 16 * 64
*   4 * 4 * 128
*   (3 * 3 * 3 * 32) + (32 * 3 * 3 * 64)


```python
#TODO define model 2
class MCNN(nn.Module):
    def __init__(self):
        super(MCNN, self).__init__()
        # Convolution 1: (3,3) kernel; padding=1; 32 filters
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        # Batchnorm 1
        self.batch1 = nn.BatchNorm2d(32)
        # ReLU Activation Layers
        self.relu = nn.ReLU()
        # Convolution 2: (3,3) kernel; padding=1; 64 filters
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        # Max-Pooling
        self.max_pool = nn.MaxPool2d(kernel_size=2)

        # Convolution 3: (3,3) kernel; padding=1; 128 filters
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1) 
        # Batchnorm 2 
        self.batch2 = nn.BatchNorm2d(128)
        # Convolution 4: (3,3) kernel; padding=1; 64 filters
        self.conv4 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        # Average-Pooling
        self.avg_pool = nn.AvgPool2d(kernel_size=16)

        # A linear layer with output of 10.
        self.fc = nn.Linear(128, 10)


    # format: N x C x W x H
    # input: 256 x 3 x 32 x 32 

    # CONV LAYER BLOCK 1
    # after conv1: 256 x 32 x 32 x 32
    # after batch1: 256 x 32 x 32 x 32
    # after relu: 256 x 32 x 32 x 32
    # after conv2: 256 x 64 x 32 x 32
    # after relu: 256 x 64 x 32 x 32
    # after max_pool: 256 x 64 x 16 x 16

    # CONV LAYER BLOCK 2
    # after conv3: 256 x 128 x 16 x 16
    # after batch2: 256 x 128 x 16 x 16
    # after relu: 256 x 128 x 16 x 16
    # after conv4: 256 x 128 x 16 x 16
    # after relu: 256 x 128 x 16 x 16
    # after avg_pool: 256 x 128 x 1 x 1
    
    # FULLY-CONNECTED BLOCK LAYER
    # after view: 256 x 128
    # after fc: 256 x 10


    def forward(self, x): 
        # CONV LAYER BLOCK 1
        x = self.relu(self.batch1(self.conv1(x)))
        x = self.max_pool(self.relu(self.conv2(x)))

        # CONV LAYER BLOCK 2
        x = self.relu(self.batch2(self.conv3(x)))
        x = self.avg_pool(self.relu(self.conv4(x)))
        
        # FULLY-CONNECTED BLOCK LAYER
        x = x.view(-1, 128*1*1)  # flattens the image
        out = self.fc(x)
        return out
```

## 5. ResNet architecture

"*Deeper neural networks are more difficult to train.  We present a residual learning framework to ease the training of networks that are substantially deeper than those used previously.*" [ https://arxiv.org/pdf/1512.03385.pdf]

That was the intention of He et al. to develop residual networks.

A ResNet is built of blocks which are called the *Residual Blocks*. 
A residual block uses *skip connections* - shortCuts â€” to jump over some layers to skip some activations.
The Resnet skips connections via addition.

**Hints**

- The downsampling in this ResNet happens between layer 1 and layer 2. In this step, the channel size is increased (32 -> 64) and the image resolution is downsampled by strided convolution (stride=2 in the Conv2d layers).
- For the adding part, the dimensions obviously have to be the same. This is accomplished by using stride=2 in the first Conv2d layer of the block while the second layer does not change the dimensions.
- The average-pooling layer has kernel size 7.
- For you to ensure everything works fine, here are the shapes in the form [(batchsize), x, y, channel] where I discard batchsize, because it does not change.
32x32x3 -> conv 7x7 -> 14x14x32 -> layer 1 -> 14x14x32 -> layer2 -> 7x7x64 -> avpool(7) -> 1x1x64 -> linear(64, 10)

```python
# TODO write a class ResidualBlock(nn.Module) where you define the blocks used in the ResNet
class ResidualBlock(nn.Module):
    def __init__(self, in_channels, out_channels, identity_downsample=None, x=1): # x = 2 if downsample, else x = 1
        super(ResidualBlock, self).__init__()
        out_channels *= x
        self.conv1 = nn.Conv2d(in_channels, out_channels, kernel_size=3, stride=x, padding=1)
        self.bn1 = nn.BatchNorm2d(out_channels)
        self.relu = nn.ReLU()
        self.conv2 = nn.Conv2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(out_channels)

        # downsample
        self.identity_downsample = identity_downsample

    def forward(self, x):
        identity = x

        x = self.relu(self.bn1(self.conv1(x)))
        x = self.bn2(self.conv2(x))
        
        # adding part
        if self.identity_downsample is not None:
            identity = self.identity_downsample(identity)
        
        x += identity
        x = self.relu(x)
        return x
```

```python
# TODO write a class ResNet(nn.Module) where you define the model
class ResNet(nn.Module):    
    def __init__(self, block, image_channels, num_classes):
        super(ResNet, self).__init__()
        self.conv1 = nn.Conv2d(image_channels, 32, kernel_size=7, stride=2, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.relu = nn.ReLU()
        self.maxpool = nn.MaxPool2d(kernel_size=3, stride=2, padding=1)

        # ResNet layers
        self.layer1 = self._make_layer(block, in_channels=32, out_channels=32)
        self.layer2 = self._make_layer(block, in_channels=32, out_channels=32, down=True)
        self.layer3 = self._make_layer(block, in_channels=64, out_channels=64)
        self.layer4 = self._make_layer(block, in_channels=64, out_channels=64) 

        self.avg_pool = nn.AvgPool2d(kernel_size=7)
        self.fc = nn.Linear(1*1*64, 10) # 1 x 1 x 64 (at least in theory)

    def forward(self, x):
        x = self.relu(self.bn1(self.conv1(x)))

        # Layer 1, Block 1 => 32 output channels
        x = self.layer1(x)
        # Layer 1, Block 2
        x = self.layer2(x)
        # Layer 2, Block 1  => 64 output channels
        x = self.layer3(x)
        # Layer 2, Block 2
        x = self.layer4(x)
        
        x = self.avg_pool(x)
        x = x.reshape(x.shape[0], -1) 
        x = self.fc(x)
        return x
        
    
    def _make_layer(self, block, in_channels, out_channels, down=False):
        identity_downsample = None
        layers = []
        
        if down:
            identity_downsample = nn.Sequential(nn.Conv2d(in_channels, out_channels*2, kernel_size=3, stride=2, padding=1),
                                                nn.BatchNorm2d(out_channels*2))
            layers.append(block(in_channels, out_channels, identity_downsample, x=2))
        else:
            layers.append(block(in_channels, out_channels, identity_downsample))

        return nn.Sequential(*layers)
```

```python
# TODO change run_training (you can copy the code from utils.py here and create a new method here 
# or just change/create a new method it in the python file)
def run_training(model, optimizer, scheduler, loss_function, device, num_epochs, 
                train_dataloader, val_dataloader, verbose=False):
    """Run model training.

    Args:
        model (nn.Module): Torch model to train
        optimizer: Torch optimizer object
        loss_fn: Torch loss function for training
        device (torch.device): Torch device to use for training
        num_epochs (int): Max. number of epochs to train
        train_dataloader (DataLoader): Torch DataLoader object to load the
            training data
        val_dataloader (DataLoader): Torch DataLoader object to load the
            validation data
        early_stopper (EarlyStopper, optional): If passed, model will be trained
            with early stopping. Defaults to None.
        verbose (bool, optional): Print information about model training. 
            Defaults to False.

    Returns:
        list, list, list, list, torch.Tensor shape (10,10): Return list of train
            losses, validation losses, train accuracies, validation accuracies
            per epoch and the confusion matrix evaluated in the last epoch.
    """
    start_time = time.time()
    master_bar = fastprogress.master_bar(range(num_epochs))
    train_losses, val_losses, train_accs, val_accs = [],[],[],[]

    for _ in master_bar:
        # Train the model
        epoch_train_loss, epoch_train_acc = utils.train(train_dataloader, optimizer, model, 
                                                  loss_function, device, master_bar)

        # Validate the model
        epoch_val_loss, epoch_val_acc, confusion_matrix = utils.validate(val_dataloader, 
                                                                   model, loss_function, 
                                                                   device, master_bar)

        # Scheduler
        if type(scheduler) is ReduceLROnPlateau:
            scheduler.step(epoch_val_loss)
        else:
            scheduler.step()

        # Save loss and acc for plotting
        train_losses.append(epoch_train_loss)
        val_losses.append(epoch_val_loss)
        train_accs.append(epoch_train_acc)
        val_accs.append(epoch_val_acc)
        
        if verbose:
            master_bar.write(f'Train loss: {epoch_train_loss:.2f}, val loss: {epoch_val_loss:.2f}, train acc: {epoch_train_acc:.3f}, val acc {epoch_val_acc:.3f}')
            
    time_elapsed = np.round(time.time() - start_time, 0).astype(int)
    print(f'Finished training after {time_elapsed} seconds.')
    return train_losses, val_losses, train_accs, val_accs, confusion_matrix
```

```python
# TODO train and plot with step size 1, gamma 0.1
epochs = 10
lr = 0.1
model = ResNet(ResidualBlock, 3, 10).cuda()
model.train()
optimizer = optim.Adam(model.parameters(), lr=lr )
scheduler= optim.lr_scheduler.StepLR(optimizer, 1, gamma=0.1) # gamme default: 0.1
train_losses, val_losses, train_accs, val_accs, confusion_matrix = run_training(
    model, optimizer, scheduler, loss_fn, device, epochs, 
                trainloader, valloader, verbose=False
)
```
